### Importación de datos



```python

import matplotlib.pyplot as plt
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

###Cambiando el tipo de dato de fecha


```python
tienda2["Fecha de Compra"] = pd.to_datetime(
    tienda2["Fecha de Compra"],
    dayfirst=True,
    errors="coerce"
)
```

```python
tienda["Fecha de Compra"] = pd.to_datetime(tienda["Fecha de Compra"])
tienda2["Fecha de Compra"] = pd.to_datetime(tienda2["Fecha de Compra"])
tienda3["Fecha de Compra"] = pd.to_datetime(tienda3["Fecha de Compra"])
```

```python
tienda4["Fecha de Compra"] = pd.to_datetime(
    tienda4["Fecha de Compra"],
    dayfirst=True
)
```

```python
tienda4["Fecha de Compra"] = pd.to_datetime(tienda4["Fecha de Compra"])
```

```python
tienda2.info()
```

###Uniendo las tiendas

```python
ventas = pd.concat([tienda, tienda2, tienda3, tienda4])
```

```python
ventas["Precio"].sum()
```

```python
tienda["Facturacion"] = tienda["Precio"] + tienda["Costo de envío"]
```

```python
tienda2["mes"] = tienda2["Fecha de Compra"].dt.month
```

```python
ingresos = {
    "Tienda ": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

print(ingresos)
```

```python
plt.bar(ingresos.keys(), ingresos.values())
plt.title("Ingresos por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")
plt.show()
```

# 2. Ventas por categoría

```python
ventas_por_categoria = ventas.groupby("Categoría del Producto")["Precio"].sum()
print(ventas_por_categoria)
```

```python
categorias = tienda["Categoría del Producto"].value_counts()

print(categorias)
```

```python
categorias.plot(kind="bar")
plt.title("Categorías más vendidas - Tienda 1")
plt.xlabel("Categoría")
plt.ylabel("Cantidad vendida")
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = ventas["Calificación"].mean()
print(calificacion_promedio)
```

```python
calificaciones = {
    "Tienda 1": tienda["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

print(calificaciones)
```

```python
plt.bar(calificaciones.keys(), calificaciones.values())
plt.title("Calificación promedio por tienda")
plt.ylabel("Calificación")
plt.show()
```

# 4. Productos más y menos vendidos

```python
productosMasVendidos= ventas["Producto"].value_counts()
print(productosMasVendidos)
```

```python
productosMenosVendidos = ventas["Producto"].value_counts().sort_values(ascending=True)
print(productosMenosVendidos)
```

```python
menos_vendidos = tienda["Producto"].value_counts().tail()

print(menos_vendidos)
```
